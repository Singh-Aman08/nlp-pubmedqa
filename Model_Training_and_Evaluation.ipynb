# Introduction

This notebook presents a comparative study of different text representation techniques and machine learning models for a multi-class text classification task (Yes / No / Maybe). We evaluate TF-IDF, Word2Vec, and BERT-based embeddings to understand their impact on performance.

These representations are tested using multiple classifiers, including Logistic Regression, SVM, XGBoost, and an Artificial Neural Network (ANN). The goal is to analyze how traditional machine learning methods compare with deep learning approaches and contextual embeddings.

Performance is measured using accuracy, precision, recall, and F1-score.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
data = np.load("embeddings_pubmedqa_train.npz", allow_pickle=True)
```

```python
data.files
```

```python
train_tfidf = data["X_tfidf"]
train_wv = data["X_wv"]
train_bert = data["X_bert"]
y_train = data["y"]
```

## Logistic Regression

This section applies Logistic Regression on TF-IDF, Word2Vec, and BERT-based embeddings to establish a baseline for text classification performance across different feature representations.

```python
log_reg_tfidf = LogisticRegression(max_iter=1000, random_state=42)
log_reg_wv = LogisticRegression(max_iter=1000, random_state=42)
log_reg_bert = LogisticRegression(max_iter=1000, random_state=42)
```

```python
log_reg_tfidf.fit(train_tfidf, y_train)
log_reg_wv.fit(train_wv, y_train)
log_reg_bert.fit(train_bert, y_train)
```

```python
data_test = np.load("embeddings_pubmedqa_test.npz", allow_pickle=True)
```

```python
test_tfidf = data_test["X_tfidf_test"]
test_wv = data_test["X_wv_test"]
test_bert = data_test["X_bert_test"]
y_test = data_test["y"]
```

```python
y_lg_tfidf_pred = log_reg_tfidf.predict(test_tfidf)
y_lg_wv_pred = log_reg_wv.predict(test_wv)
y_lg_bert_pred = log_reg_bert.predict(test_bert)
```

```python
from sklearn.metrics import classification_report

print("###############  TF-IDF EVALUATION    ###############")
print(classification_report(y_test, y_lg_tfidf_pred))

print("############### WORDTOVEC EVALUATION #################")
print(classification_report(y_test, y_lg_wv_pred))

print("############### SCIBERT EVALUATION   #################")
print(classification_report(y_test, y_lg_bert_pred))

```

```python
from sklearn.metrics import classification_report

print("###############  TF-IDF EVALUATION    ###############")
print(classification_report(y_test, y_svm_tfidf_pred))

print("############### WORDTOVEC EVALUATION #################")
print(classification_report(y_test, y_svm_wv_pred))

print("############### SCIBERT EVALUATION   #################")
print(classification_report(y_test, y_svm_bert_pred))
```

## XGBoost

This section applies XGBoost on TF-IDF, Word2Vec, and BERT-based embeddings to evaluate its performance for multi-class text classification and compare it with other machine learning models.

```python
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)
```

```python
xgb_tfidf = XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="mlogloss",
    use_label_encoder=False)
xgb_wv = XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="mlogloss",
    use_label_encoder=False)

xgb_bert = XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="mlogloss",
    use_label_encoder=False)
```

```python
xgb_tfidf.fit(train_tfidf, y_train_enc)
xgb_wv.fit(train_wv, y_train_enc)
xgb_bert.fit(train_bert, y_train_enc)
```

```python
y_xgb_tfidf_pred = xgb_tfidf.predict(test_tfidf)
y_xgb_wv_pred = xgb_wv.predict(test_wv)
y_xgb_bert_pred = xgb_bert.predict(test_bert)
```

```python
from sklearn.metrics import classification_report

print("###############  TF-IDF EVALUATION   ###############")
print(classification_report(y_test_enc, y_xgb_tfidf_pred))

print("############### WORDTOVEC EVALUATION #################")
print(classification_report(y_test_enc, y_xgb_wv_pred))

print("############### SCIBERT EVALUATION   #################")
print(classification_report(y_test_enc, y_xgb_bert_pred))
```

## Artificial Neural Network (ANN)

This section implements a feedforward Artificial Neural Network (ANN) on TF-IDF, Word2Vec, and BERT-based embeddings to evaluate the performance of deep learning for multi-class text classification.

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
train_tfidf_tensor = torch.tensor(train_tfidf, dtype=torch.float32)
train_wv_tensor = torch.tensor(train_wv, dtype=torch.float32)
train_bert_tensor = torch.tensor(train_bert, dtype=torch.float32)

test_tfidf_tensor = torch.tensor(test_tfidf, dtype=torch.float32)
test_wv_tensor = torch.tensor(test_wv, dtype=torch.float32)
test_bert_tensor = torch.tensor(test_bert, dtype=torch.float32)

y_train_tensor = torch.tensor(y_train_enc, dtype=torch.long)
y_test_tensor = torch.tensor(y_test_enc, dtype=torch.long)
```

```python
class SimpleNN(nn.Module):
    def __init__(self, input_dim, num_classes):
        super().__init__()
        self.model = nn.Sequential(nn.Linear(input_dim, 128), nn.ReLU(), nn.Dropout(0.3), nn.Linear(128, 64)
                                   ,nn.ReLU(), nn.Linear(64,num_classes))
        
    def forward(self, x):
        return self.model(x)
    
    
    
    
input_dim = train_tfidf.shape[1]
num_classes = len(set(y_train))

model = SimpleNN(input_dim, num_classes)
```

```python
def model_training(train_set, train_set_tensor, y_train_tensor):
    criterion = nn.CrossEntropyLoss()
  

    input_dim = train_set.shape[1]
    num_classes = 3
    model = SimpleNN(input_dim, num_classes)
    optimizer = optim.Adam(model.parameters(), lr = 0.001)
    epochs = 10
    for epoch in range(epochs):
        model.train()
        optimizer.zero_grad()
        outputs = model(train_set_tensor)
        loss = criterion(outputs, y_train_tensor)
        loss.backward()
        optimizer.step()
        
        print(f"Epoch {epoch+1}, Loss: {loss.item()}")
        
    return model
        
```

```python
def model_evaluation(model, test_set_tensor, y_test):

    model.eval()
    with torch.no_grad():
        output = model(test_set_tensor)
        preds  = torch.argmax(output, axis=1).numpy()
    y_pred = le.inverse_transform(preds)
    print(classification_report(y_test, y_pred))
```

```python
trained_model = model_training(train_tfidf, train_tfidf_tensor, y_train_tensor)
```

```python
model_evaluation(trained_model, test_tfidf_tensor, y_test)
```

```python
trained_model = model_training(train_wv, train_wv_tensor, y_train_tensor)
model_evaluation(trained_model, test_wv_tensor, y_test)
```
